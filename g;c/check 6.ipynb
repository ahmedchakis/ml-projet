```python
#Reading the dataset
import pandas as pd
data1=pd.read_csv("iris.csv")

#Converting categorical to numerical so we can use knn
data1['species']=data1['species'].map({'setosa':0,'versicolor':1,'virginica':2})

#Choosing the features as x and the target as y
x=data1[['sepal_length','sepal_width','petal_length','petal_width']]
y=data1['species']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=30) #split our data with test size of 20% 

knn=KNeighborsClassifier(n_neighbors=20) #build our knn classifier
knn.fit(x_train,y_train) #Training KNN classifier
y_pred=knn.predict(x_test)  #Testing
print('Accuracy=',accuracy_score(y_pred,y_test))
```

```python
n_neighbors=30
scores=[]
for k in range(1,30):
    knn=KNeighborsClassifier(n_neighbors-k)
    knn.fit(x_train,y_train)
    y_pred=knn.predict(x_test)
    print('Accuracy for k=',k,'is:',round(accuracy_score(y_pred,y_test),2))
    scores.append(round(accuracy_score(y_pred,y_test),2))
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,30),scores)
plt.xlabel('Value K for KNN')
plt.ylabel('Testing Accuracy')
```

```python
#2 Split the training set and the test set
def loadDataset(filename, split, trainingSet=[] , testSet=[]):
    with open(filename, 'r') as csvfile:

        lines = csv.reader(csvfile)

        dataset = list(lines)

        for x in range(len(dataset)-1):
            
            d=[]
            for y in range(4):
                dataset[x][y] = float(dataset[x][y])
                d.append(dataset[x][y])
            d.append(dataset[x][4])
                
            if random.random() < split:

                    #if the randomly generated float is smaller than the split value , then append the dataset value into the training set
                    trainingSet.append(d)

            else:
                    #else append the dataset value into the testing set
                    testSet.append(d)

#We can test this function out with our iris dataset, as follows:

trainingSet=[]

testSet=[]

loadDataset('iris.data.txt', 0.66, trainingSet, testSet)

print ('Train: ' + repr(len(trainingSet)))

print ('Test: ' + repr(len(testSet)) )
```

```python
#3 calculating the eucleudean distance between 2 instances
import math
def euclideanDistance(instance1, instance2, length):
    res=0
    for i in range(0,length):     
        res =res+math.pow(instance1[i]-instance2[i],2)
    distance=math.sqrt(res)
    return distance

distance=euclideanDistance([2,2,3,'a'],[3,1,5,'b'],3)
print('Distance: ' + repr(distance))
```

```python
#4 Get the k-nearest neighbors
import operator

def getNeighbors(trainingSet, testInstance, k):

    #empty list that will contain a tuple of instances from the training set and the distance from the testing set
    distances = []

    length = len(testInstance)-1

    for x in range(len(trainingSet)):
        

        dist = euclideanDistance(testInstance, trainingSet[x], length)

        distances.append((trainingSet[x], dist))
        
    
    #sorting tuples by their distances 
    distances.sort(key=operator.itemgetter(1))
        
    neighbors = []

    for x in range(k):
        

        neighbors.append(distances[x][0])

    return neighbors


trainSet = [[4, 4, 4, 'b'],[2, 2, 2, 'a'],[3, 3, 3, 'c'],[5.5, 5.5, 5.5, 'b']]

testInstance = [5, 5, 5,'e']

k = 3

neighbors = getNeighbors(trainSet, testInstance, k)

print(neighbors)
```

```python
#5 determine which category the instance belongs to 
import operator

def getResponse(neighbors):

    classVotes = {}

    for x in range(len(neighbors)):

        response = neighbors[x][len(neighbors[x])-1] #complete with appropriate number

        if response in classVotes:

            classVotes[response]=classVotes[response]+1
        else:
            classVotes[response]=1

    sortedVotes = sorted(classVotes.items(), key=operator.itemgetter(1), reverse=True)

    return sortedVotes[0][0]

#We can test out this function with some test neighbors, as follows:

neighbors = [[1,1,1,'a'], [2,2,2,'b'], [3,3,3,'b']]

response = getResponse(neighbors)

print(response)
```

```python
#6 determine the accuracy of our model
def getAccuracy(testSet, predictions):
    correct=0
    for x in range(len(testSet)):
        if (testSet[x][len(testSet[x])-1]==predictions[x]):
            correct=correct+1
    
    return (correct/float(len(testSet))) * 100.0

#We can test this function with a test dataset and predictions, as follows:

testSet = [[1,1,1,'a'], [2,2,2,'a'], [3,3,3,'b'],[4,4,4,'b']]

predictions = ['a', 'a', 'a','b']

accuracy = getAccuracy(testSet, predictions)

print(accuracy)
```

```python
#      ˜”*°•.˜”*°•☆ Main ☆•°*”˜.•°*”˜ 
trainingSet=[]
testSet=[]
loadDataset('iris.data.txt', 0.66, trainingSet, testSet)
print ('Train: ' + repr(len(trainingSet)))
print ('Test: ' + repr(len(testSet)) )
k=29
predictions=[]
for x in testSet:
    n=getNeighbors(trainingSet,x,k)
    predictions.append(getResponse(n))


accuracy = getAccuracy(testSet, predictions)

print('Accuracy =',accuracy)
```
